```python
import os
from PIL import Image
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np


images = []
def import_jpg_files(folder_path):
    if not os.path.exists(folder_path):
        print("Path does not exist")
        return

    for file_name in os.listdir(folder_path):
        image_raw = imread(os.path.join(folder_path, file_name))
        # convert the images into gray mode
        image_sum = image_raw.sum(axis=2)
        image_bw = (image_sum/image_sum.max()*255).astype(np.uint8)   # black and white
        # image_array = list(image_gray.getdata())
        
        # single_image_array = [image_array[pixel_no: pixel_no+width] for pixel_no in range(0, len(image_array), width)]
        images.append(image_bw)
    # Displaying the image
    width, height = image_bw.shape
    plt.figure(figsize=[4,6])
    plt.imshow(image_bw, cmap=plt.cm.gray)
    print(image_bw.shape)
    return width, height

# 
dir = "C:\\Users\\27615\\Desktop\\\\2024.2\\3790 1\\assignments\\7\\FEI Face Database\\frontalimages_manually_aligned"
width, height = import_jpg_files(dir)
display(images)
print(len(images))
print(width, height)

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Get de-mean faces
flattened_demean_faces = [flattened_image - flattened_mean_face for flattened_image in flattened_images]

# Create DataFrame for de-mean faces
df_demean = pd.DataFrame()

# Store de-mean faces in DataFrame
for dm_no, de_mean_face in enumerate(flattened_demean_faces):
    df_demean[dm_no] = de_mean_face

# Display the DataFrame
display(df_demean)

# Show the de-mean faces
for i_no in range(0, 400, 100):
    demean_matrix = np.reshape(flattened_demean_faces[i_no], (width, height))  # 
    plt.imshow(demean_matrix.astype(np.float64), cmap=plt.cm.gray)  
    plt.show()
```

```python
from sklearn.decomposition import PCA
train_data = df_demean.T.values
display(train_data.shape)

# PCAn
pca = PCA(n_components=399)

# PCA
pca.fit(train_data)

# PCA
mean_ = pca.mean_
components_ = pca.components_
explained_variance_ = pca.explained_variance_

# 
reduced_train_data = pca.transform(train_data)

# 
print(reduced_train_data.shape)
display(reduced_train_data)

df_reduced_demean = pd.DataFrame(reduced_train_data.T)
display(df_reduced_demean)
```

```python
# calculate covariance
num_columns = df_reduced_demean.shape[1]
sum_tensor_products = 0
for col in range(num_columns):
    single_tensor_product = np.outer(df_reduced_demean[col], df_reduced_demean[col])
    sum_tensor_products += single_tensor_product
    
covariance = sum_tensor_products / num_columns

df_covariance = pd.DataFrame(covariance)
display(df_covariance)
```

```python
# get the eigen-value and eigen-vector
eig_values, eig_vectors = np.linalg.eig(covariance)

# Sort eigenvalues in descending order and get the indices
sorted_indices = np.argsort(eig_values)[::-1]
# Select the top 10 eigenvalues and eigenvectors
largest_10_indices = sorted_indices[:10]
top_10_eig_vectors = eig_vectors[:, largest_10_indices]

# Store the top 10 eigenvectors in a list B
B = [top_10_eig_vectors[:, i].tolist() for i in range(top_10_eig_vectors.shape[1])]

# Print the top 10 eigenvalues
df_10_eigvects = pd.DataFrame(top_10_eig_vectors)
display(df_10_eigvects)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# get the minimum of each difference between inner product of demean faces*eigenvector and input*eigenvector
def find_min_columns(df_flattened_conv_dm, reduced_input_flattened_matrix, B, top_n=5):
    # 
    euclidean_dists = []
    
    for col in range(df_flattened_conv_dm.shape[1]):
        sum_dist = 0
        for single_eigvect in B:
            single_col_data = df_flattened_conv_dm[col]
            # 
            squared_diff_inner_product = (np.dot(single_col_data, single_eigvect) - np.dot(reduced_input_flattened_matrix[0], single_eigvect))**2
            # 10eigenvects
            sum_dist += squared_diff_inner_product
        # 
        euclid_dist = np.sqrt(sum_dist)
        euclidean_dists.append(euclid_dist)
    
    # top_n
    top_n_indices = np.argsort(euclidean_dists)[:top_n]
    
    return top_n_indices, [euclidean_dists[i] for i in top_n_indices]

# 
min_columns_indices, min_dists = find_min_columns(df_reduced_demean, df_reduced_input, B, top_n=5)
print("Closest training pictures to the input picture:")
print("Top 5 closest image indices:", min_columns_indices)

# 5
closest_images = [images[i] for i in min_columns_indices]

# 
fig, axes = plt.subplots(1, 5, figsize=(15, 5))
for ax, img, dist, idx in zip(axes, closest_images, min_dists, min_columns_indices):
    ax.imshow(img, cmap=plt.cm.gray)
    ax.set_title(f"Index: {idx}\nDist: {dist:.2f}")
    ax.axis('off')
plt.show()
```

```python
import cv2
import copy
import numpy as np

# 
train_images = images

# 
test_image = input_image_bw

# Haar
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# test_image
test_image_copy = copy.deepcopy(test_image)

faces = face_cascade.detectMultiScale(test_image_copy, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# 
for (x, y, w, h) in faces:
    cv2.rectangle(test_image_copy, (x, y), (x+w, y+h), (255, 0, 0), 2)

# 
# cv2.imshow('Test Image', test_image_copy)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# 
def histogram_similarity(hist1, hist2):
    return cv2.compareHist(hist1, hist2, cv2.HISTCMP_CORREL)

# 
test_face_histogram = cv2.calcHist([test_image], [0], None, [256], [0, 256])

# 
similarities = []
for train_image in train_images:
    train_face_histogram = cv2.calcHist([train_image], [0], None, [256], [0, 256])
    similarity = histogram_similarity(test_face_histogram, train_face_histogram)
    similarities.append(similarity)

# 5
top_5_indices = np.argsort(similarities)[-5:][::-1]   # 5, 
top_5_images = [train_images[i] for i in top_5_indices]

# 5
# for idx, img in enumerate(top_5_images):
#     cv2.imshow(f'Similar Image {idx+1}', img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# Create a dictionary to store the sum of occurrences of each index
indexSum_dict = {}
# Iterate over both lists and count occurrences
for index_rank in range(len(min_columns_indices)):
    indexSum_dict[min_columns_indices[index_rank]] = index_rank

for index_rank in range(len(top_5_indices)):
    if top_5_indices[index_rank] in indexSum_dict:
        indexSum_dict[top_5_indices[index_rank]] += index_rank
    else:
        indexSum_dict[top_5_indices[index_rank]] = index_rank

indexCount_dict = {}
for index in min_columns_indices:
    indexCount_dict[index] = 1
for index in top_5_indices:
    if index in indexCount_dict:
        indexCount_dict[index] += 1
    else:
        indexCount_dict[index] = 1

display(indexSum_dict)
display(indexCount_dict)
# Find the index with the minimum sum of occurrences
# Find the index with the maximum count
max_count_indices = [index for index, count in indexCount_dict.items() if count == max(indexCount_dict.values())]

most_similar_index = None
# If there is only one index with maximum count, set it as the most similar index
if len(max_count_indices) == 1:
    most_similar_index = max_count_indices[0]
else:
    # Otherwise, find the index with the minimum sum of occurrences among those with maximum count
    min_sum_index = min(max_count_indices, key=indexSum_dict.get)
    most_similar_index = min_sum_index

print("Most similar index:", most_similar_index)

# display the result image
closest_image = images[most_similar_index]
display(closest_image)
plt.imshow(closest_image, cmap=plt.cm.gray)
plt.show()
```
