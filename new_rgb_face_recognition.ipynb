```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

images = []
def import_jpg_files(image_list, folder_path):
    if not os.path.exists(folder_path):
        print("Path does not exist")
        return

    for file_name in os.listdir(folder_path):
        image = Image.open(os.path.join(folder_path, file_name))
        width, height = image.size
        image_data = np.array(image)
        # image_data = list(image.getdata())
        # single_image_array = [image_data[pixel_no: pixel_no+width] for pixel_no in range(0, len(image_data), width)]
        images.append(image_data)
    return width, height
        
# 调用函数并传入文件夹路径
dir = "C:\\Users\\27615\\Desktop\\温肯课程\\这是2024.2开始的\\数学机器学习3790 1班\\我的assignments\\作业7\\FEI Face Database\\frontalimages_manually_aligned"
width, height = import_jpg_files(images, dir)

display(images)
print(len(images))
print(width, height)

```

```python
import numpy as np
import pandas as pd
# flatten the images
flattened_images = [np.array(image).ravel() for image in images]
# store the flattened images in a .csv file
# Convert flattened_images to DataFrame
df_jpgs = pd.DataFrame()
for image_no in range(len(flattened_images)):
    df_jpgs[image_no] = flattened_images[image_no]

display(df_jpgs)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 计算 "Mean Face"
sum_face = np.zeros_like(flattened_images[0], dtype=np.int64)  # 创建一个和第一张图片大小相同的全零数组
for image_No in range(len(flattened_images)):
    sum_face = sum_face + flattened_images[image_No]

flattened_mean_face = sum_face / len(flattened_images)
print(flattened_mean_face)
# 将 flattened_mean_face 转换为 numpy 数组
flattened_mean_face = np.array(flattened_mean_face)

# 将 mean_face 按照每3个像素一组重新组织
mean_face = np.reshape(flattened_mean_face, (-1, 3))

# 保存 mean_face 
df_mean = pd.DataFrame(mean_face)

# print("CSV file saved successfully at:", csv_file_path)
display(df_mean)

# 显示平均脸
meanface_matrix = np.reshape(mean_face, (height, width, 3))  # 将 mean_face 转换为矩阵形式
plt.imshow(meanface_matrix.astype(np.uint8))  # 注意要将数据类型转换为 uint8
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Get de-mean faces
flattened_demean_faces = [flattened_image - flattened_mean_face for flattened_image in flattened_images]

# Create DataFrame for de-mean faces
df_demean = pd.DataFrame()

# Store de-mean faces in DataFrame
for dm_no, de_mean_face in enumerate(flattened_demean_faces):
    df_demean[dm_no] = de_mean_face

# Display the DataFrame
display(df_demean)

# Show the de-mean faces
for i_no in range(0, 400, 100):
    demean_matrix = np.reshape(flattened_demean_faces[i_no], (height, width, 3))  # 将去均值化的图像转换为矩阵形式
    plt.imshow(demean_matrix.astype(np.float64))  
    plt.show()
```

```python
# Assign a total
df_weighted_demean = pd.DataFrame()

# Store de-mean faces in DataFrame
for dm_no, de_mean_face in enumerate(flattened_demean_faces):
    # 计算加权平均，每3个元素（对应RGB）加权成1个元素
    weighted_de_mean_face = np.zeros(len(de_mean_face) // 3)  # 初始化加权后的数组
    for i in range(0, len(de_mean_face), 3):
        # 按照一定的权重对每个像素的 RGB 值进行加权平均
        weighted_pixel = 1 * de_mean_face[i] + 1 * de_mean_face[i+1] + 1 * de_mean_face[i+2]
        # 将加权后的结果存入加权后的数组
        weighted_de_mean_face[i // 3] = weighted_pixel
        
    # 将加权后的 de_mean_face 赋值给 df_weighted_demean[dm_no]
    df_weighted_demean[dm_no] = weighted_de_mean_face
    
display(df_weighted_demean)
```

```python
from sklearn.decomposition import PCA

# 定义PCA模型并拟合训练集数据
pca = PCA(n_components=399)  # 设定降维后的维度
pca.fit(df_weighted_demean.T)  # df_weighted_demean.T是为了符合sklearn的输入要求，即样本在行上，特征在列上

# 对训练集和测试集进行降维
reduced_demean_data = pca.transform(df_weighted_demean.T)

# 将降维后的数据转换为DataFrame
df_reduced_demean = pd.DataFrame(reduced_demean_data.T)

# 显示降维后的数据
display(df_reduced_demean)
```

```python
# calculate covariance
num_columns = df_reduced_demean.shape[1]
sum_tensor_products = 0
for col in range(num_columns):
    single_tensor_product = np.outer(df_reduced_demean[col], df_reduced_demean[col])
    sum_tensor_products += single_tensor_product
    
covariance = sum_tensor_products / num_columns

df_covariance = pd.DataFrame(covariance)
display(df_covariance)
```

```python
# get the eigen-value and eigen-vector
eig_values, eig_vectors = np.linalg.eig(covariance)

# Sort eigenvalues in descending order and get the indices
sorted_indices = np.argsort(eig_values)[::-1]
# Select the top 10 eigenvalues and eigenvectors
largest_10_indices = sorted_indices[:10]
top_10_eig_vectors = eig_vectors[:, largest_10_indices]

# Store the top 10 eigenvectors in a list B
B = [top_10_eig_vectors[:, i].tolist() for i in range(top_10_eig_vectors.shape[1])]

# Print the top 10 eigenvalues
df_10_eigvects = pd.DataFrame(top_10_eig_vectors)
display(df_10_eigvects)
```
